#Cancer Prediction

# Objective
To predict whether a tumor is malignant (M) or benign (B) based on the features of cell nuclei from a breast mass.

```python
#STEP 1: Importing Libraries
import pandas as pd
import numpy as np
```

```python
#STEP 2: dataset (Ybi foundation)
data = pd.read_csv("https://github.com/YBIFoundation/Dataset/raw/main/Cancer.csv")
```

```python
#STEP 3: Knowing about data
data.head()
data.tail()
data.info()
data.describe()
data.shape
data.columns
```

```python
#STEP 4: Feature Variable and Target Variable
x = data.drop(['id','diagnosis','Unnamed: 32'],axis=1)
y = data['diagnosis']
```

```python
#STEP 5: Train-Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.2, random_state=42)
x_train.shape,x_test.shape
```

```python
#STEP 6: MODELING
# Logistic Regression
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression(max_iter=6000)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
from sklearn.svm import SVC
svm_model = SVC(probability=True, random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_model = GradientBoostingClassifier(n_estimators=100, random_state=42)
```

```python
#STEP 7 : Model Training
log_model.fit(x_train,y_train)
```

```python
rf_model.fit(x_train,y_train)
```

```python
dt_model.fit(x_train,y_train)
```

```python
knn_model.fit(x_train,y_train)
```

```python
svm_model.fit(x_train,y_train)
```

```python
gb_model.fit(x_train,y_train)
```

```python
log_model.intercept_
log_model.coef_
```

```python
y_pred1 = log_model.predict(x_test)
print(y_pred1)
```

```python
y_pred2 = rf_model.predict(x_test)
print(y_pred2)
```

```python
y_pred3 = dt_model.predict(x_test)
print(y_pred3)
```

```python
y_pred4 = svm_model.predict(x_test)
print(y_pred4)
```

```python
y_pred5 = knn_model.predict(x_test)
print(y_pred1)
```

```python
y_pred6 = gb_model.predict(x_test)
print(y_pred6)
```

```python
#STEP 8 : Model Accuracy
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
models = {
    "Logistic Regression": log_model,
    "Random Forest": rf_model,
    "Decision Tree": dt_model,
    "K-Nearest Neighbors": knn_model,
    "SVM": svm_model,
    "Gradient Boosting": gb_model
}

# Evaluate each model
for name, model in models.items():
    y_pred = model.predict(x_test)
    print(f"--- {name} Results ---")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("\n")
```

# Explanation
 We predicted whether a breast tumor is malignant or benign using features of cell nuclei like radius, texture, area, smoothness, and others. We trained models such as Logistic Regression, Decision Tree, Random Forest, KNN, SVM, and Gradient Boosting, and evaluated them using accuracy and confusion matrix. The models achieved high accuracy (around 97%)
